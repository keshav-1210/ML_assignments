```python
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_column(data, column='median_house_value', x_col='median_income'):

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))


    sns.kdeplot(data[column], ax=axes[0])
    axes[0].set_title(f'KDE of {column}')
    axes[0].set_xlabel(column)

    sns.boxplot(x=data[column], ax=axes[1])
    axes[1].set_title(f'Boxplot of {column}')
    axes[1].set_xlabel(column)

    sns.scatterplot(x=data[x_col], y=data[column], ax=axes[2])
    axes[2].set_title(f'Scatterplot of {x_col} vs {column}')
    axes[2].set_xlabel(x_col)
    axes[2].set_ylabel(column)


    plt.tight_layout()
    plt.show()

visualize_column(housing)
```

```python
def identify_and_visualize_outliers(housing,column_name):
    #95% mean +-2 standard deviation
    mean = housing[column_name].mean()
    std_dev = housing[column_name].std()

    lower_bound = mean - 2 * std_dev
    upper_bound = mean + 2 * std_dev

    outliers = housing[(housing[column_name] < lower_bound) | (housing[column_name] > upper_bound)]
    num_outliers = outliers.shape[0]

    print(f"Total number of outliers in '{column_name}': {num_outliers}")

    plt.figure(figsize=(10, 6))

    sns.kdeplot(housing[column_name], fill=True, color='skyblue', label='Data Distribution')

    plt.axvspan(housing[column_name].min(), lower_bound, color='red', alpha=0.3, label='Outlier Region (Low)')
    plt.axvspan(upper_bound, housing[column_name].max(), color='red', alpha=0.3, label='Outlier Region (High)')

    plt.axvline(mean, color='black', linestyle='--', label='Mean')
    plt.axvline(lower_bound, color='orange', linestyle='--', label='Lower Bound (-2)')
    plt.axvline(upper_bound, color='orange', linestyle='--', label='Upper Bound (+2)')

    plt.title(f'Data Distribution of {column_name} with Outliers Highlighted', fontsize=16)
    plt.xlabel(column_name, fontsize=14)
    plt.ylabel('Density', fontsize=14)

    plt.legend()

    plt.show()
    print(lower_bound)
    print(std_dev)
identify_and_visualize_outliers(housing,'median_house_value')
```
